```python
import numpy as np
import pandas as pd
```

```python
#Dataset collection
url = str(input("Enter the url or relative path for the dataset:\n"))
# Datasets/california_housing_train.csv
dataset = pd.read_csv(url)
```

# Data Cleaning and Seperation

```python
#Identifying dependent variables
features = dataset.columns
for ind,name in enumerate(features):
    print(f"{ind} : {name}")
while(True):
    try:
        dependent = features[int(input("\nChoose the dependent variables: "))]
        break
    except ValueError:
        print("Invalid Datatype. Please enter int.")
    except IndexError:
        print(f"Invalid Number. select from 0 to {len(features)-1}")
print(f"\"{dependent}\" is the dependent variable")
```

```python
# Seperating dependent and independent variables
x_param = dataset.drop(dependent,axis = 1)
y_param = pd.DataFrame(dataset[dependent])

for ind,name in enumerate(x_param.columns):
    print(f"{ind} : {name}")
while(True):
    try:
        useful_param = set(map(int,input("\nChoose the dependent variables: ").split()))
        useful_param = [x_param.columns[col] for col in useful_param]
        break
    except ValueError:
        print("Invalid Datatype. Please enter int.")
    except IndexError:
        print(f"Invalid Number. select from 0 to {len(x_param.columns)-1}")
x_param = pd.DataFrame(x_param[useful_param])
```

# Linear Regression

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
lin_mod = LinearRegression()
lin_mod.fit(x_param,y_param)
```

```python
test_data = pd.read_csv("Datasets/california_housing_test.csv")
```

```python
x_test,y_test = pd.DataFrame(test_data[useful_param]),pd.DataFrame(test_data[dependent])
```

```python
y_test
```

```python
y_pred = lin_mod.predict(x_test)
y_pred
```

```python
results = pd.DataFrame(y_pred,columns=["predicted"])
results = results.assign(actual=y_test)
results
```

```python
import cv2
import numpy as np
import pandas as pd
from tkinter import filedialog as fd

#image location acquiring through tkinter filedialog
img_dir = ''
while(not img_dir):
    img_dir = fd.askopenfilename(title="Select Image",
                                  filetypes=(('jpg files','.jpg'),
                                             ('png files','.png'),
                                             ('jpeg files','.jpeg'),
                                             ('bmp images','.bmp'),
                                             ('all types','*.*')))
# open image with opencv
img = cv2.imread(img_dir)

Pixel_select = False
r = g = b = xpos = ypos = 0

#import colors.csv file
index=["color","color_name","hex","R","G","B"]
colors = pd.read_csv('/home/dinesh/projects/codeclause/colors.csv', names=index, header=None)

#identifying color name through differences between RGB values
def ColorName(R,G,B):
    minimum = 10000
    for i in range(len(colors)):
    	color = colors.iloc[i]
    	diff = abs(R- int(color["R"])) + abs(G- int(color["G"]))+ abs(B- int(color["B"]))
    	if(diff<=minimum):
    	    minimum = diff
    	    color_name = color["color_name"]
    return color_name

#function to get x,y coordinates of mouse double click
def pixel_data(event, x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b,g,r,xpos,ypos, Pixel_select
        Pixel_select = True
        xpos = x
        ypos = y
        b,g,r = img[y,x]
        b = int(b)
        g = int(g)
        r = int(r)
       
cv2.namedWindow('DoubleClick to pick Color (Esc to exit)')
cv2.setMouseCallback('DoubleClick to pick Color (Esc to exit)',pixel_data)

while(True):
    cv2.imshow('DoubleClick to pick Color (Esc to exit)',img)
    if (Pixel_select):
   	# display color details
        cv2.rectangle(img,(20,20), (800,50), (b,g,r), -1)
        text = ColorName(r,g,b) + ' R= '+ str(r) +  ' G= '+ str(g) +  ' B= '+ str(b)
        cv2.putText(img, text,(40,42),2,0.7,(255,255,255),1,cv2.LINE_AA)
        if(r+g+b>=600):
            cv2.putText(img, text,(40,42),2,0.7,(0,0,0),1,cv2.LINE_AA)            
        Pixel_select=False


    if cv2.waitKey(20) & 0xFF ==27:
        break
    
cv2.destroyAllWindows()
```
